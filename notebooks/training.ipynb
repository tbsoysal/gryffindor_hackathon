```python
def get_total_rooms(val):
    val = str(val).lower().strip()

    if 'studio' in val:
        return 1.0
    elif '10 and more' in val:
        return 12.0

    try:
        if '+' in val:
            parts = val.split('+')
            return float(parts[0]) + float(parts[1])
    except:
        return np.nan
    return np.nan

df['Number of rooms'] = df['Number of rooms'].apply(get_total_rooms)
    
```

```python
def conver_age_to_numeric(val):
    val = str(val).lower().strip()

    if 'between' in val:
        clean_val = val.replace('between', '').strip()
        try:
            low, high = clean_val.split('-')
            return (float(low) + float(high)) / 2
        except:
            return np.nan
    elif 'more than' in val:
        return 35.0
    else:
        try:
            return float(val)
        except:
            return np.nan

df['Building Age'] = df['Building Age'].apply(conver_age_to_numeric)
```

```python
def convert_floor_to_numeric(val):
    val = str(val).lower().strip()
    
    # 1. Kot Daireler (Negatif Katlar)
    if 'kot' in val:
        try:
            # "kot 2" -> 2'yi alıp eksi yapıyoruz -> -2
            number = float(val.split()[1])
            return -number
        except:
            return -1.0 # Okunamazsa varsayılan -1
            
    # 2. Bodrum (Basement)
    elif 'basement' in val:
        return -1.0
    
    # 3. Zemin ve Giriş Katlar (0 kabul edilir)
    elif any(x in val for x in ['ground', 'entrance', 'garden', 'villa', 'private']):
        return 0.0
    
    # 4. Yüksek Giriş (Zeminden az daha değerli)
    elif 'high entrance' in val:
        return 0.5
    
    # 5. En Üst Katlar / Penthouse / 30 üzeri
    elif 'penthouse' in val or 'more than' in val:
        return 35.0 # Veri setindeki mantıklı bir tavan değer
    
    # 6. Normal Sayısal Katlar
    else:
        try:
            return float(val)
        except:
            return np.nan

df['Floor location'] = df['Floor location'].apply(convert_floor_to_numeric)
```

```python
def convert_total_floors(val):
    val = str(val).lower().strip()
    
    # "30 and more than" durumu
    if 'more than' in val:
        # 30'dan fazla dediği için ortalama bir gökdelen/rezidans katı olarak 35 veya 40 verebiliriz.
        return 35.0 
    
    # Normal sayılar
    try:
        return float(val)
    except:
        return np.nan

df['Number of floors'] = df['Number of floors'].apply(convert_total_floors)
```

```python
def simplify_heating(val):
    val = str(val).lower().strip()
    
    # 1. Kombi (En yaygın)
    if 'combi' in val:
        return 'Kombi'
    
    # 2. Yerden Isıtma (Değerli bir özellik)
    elif 'floor' in val: # underfloor ve floor heating'i yakalar
        return 'Yerden Isıtma'
    
    # 3. Merkezi Sistemler
    elif 'central' in val or 'center' in val:
        return 'Merkezi Sistem'
    
    # 4. Klima ve Elektrikli modern sistemler
    elif any(x in val for x in ['climate', 'fancoil', 'vrv', 'heat pump']):
        return 'Klima/Elektrikli'
    
    # 5. Soba veya Isıtma Yok (Fiyatı düşüren unsurlar)
    elif 'stove' in val or 'absent' in val or 'fireplace' in val:
        return 'Soba/Yok'
    
    # 6. Diğer (Jeotermal, Solar vb.)
    else:
        return 'Diğer'

df['Simplified heating group'] = df['Heating'].apply(simplify_heating)
print("Temizlenmis Gruplar:")
print(df['Simplified heating group'].value_counts())
```

```python
def clean_price(val):
    val = str(val).strip()
    if '...' in val or val == '':
        return np.nan
    # 1. TL, Sterlin, Dolar vb. sembolleri ve harfleri temizle
    val = val.replace('TL', '').replace('tl', '')
    
    # 2. Noktayı (.) tamamen sil (Binlik ayıracı olduğu için)
    # Örnek: 1.615.000 -> 1615000 olur.
    val = val.replace('.', '')
    
    # 3. Virgül (,) varsa ondalıktır, onu noktaya çevir (Nadir ama olabilir)
    val = val.replace(',', '.')
    
    # 4. Boşlukları temizle ve float'a çevir
    try:
        return float(val.strip())
    except:
        return np.nan

df['Price'] = df['Price'].apply(clean_price)
```

```python
print("Model eğitiliyor... (Bu işlem bilgisayar hızına göre 1-2 dakika sürebilir)")
rf_model.fit(X_train, y_train)
```

```python
from sklearn.model_selection import GridSearchCV

# 1. Denenecek ayarların listesi (Menü)
parametre_gridi = {
    'n_estimators': [100, 200, 300],        # Ağaç sayısı seçenekleri
    'max_depth': [None, 10, 20, 30],        # Derinlik seçenekleri
    'min_samples_split': [2, 5, 10],        # Bölünme kuralı
    'min_samples_leaf': [1, 2, 4]           # Yaprak kuralı
}

# 2. Modeli ve GridSearch'ü hazırla
rf = RandomForestRegressor(random_state=42)

# cv=3: Veriyi 3 parçaya bölüp çapraz doğrulama yapar (Sağlamasını yapar)
grid_search = GridSearchCV(estimator=rf, param_grid=parametre_gridi, 
                           cv=3, n_jobs=-1, verbose=2)

print("En iyi ayarlar aranıyor... Kahveni al bekle ☕")
grid_search.fit(X_train, y_train)

# 3. Sonuçları Göster
print("\nEn İyi Ayarlar:", grid_search.best_params_)

# 4. En iyi ayarlarla skoru tekrar ölç
best_model = grid_search.best_estimator_
y_pred_new = best_model.predict(X_test)
print(f"Yeni R2 Skoru: {r2_score(y_test, y_pred_new):.4f}")
```

```python
# Gerekli kütüphaneler
from sklearn.ensemble import VotingRegressor, RandomForestRegressor, GradientBoostingRegressor
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
import xgboost as xgb # XGBoost kütüphanesi

# 1. TAKIM OYUNCULARINI SEÇELİM
# Random Forest (Genel kaptan)
rf = RandomForestRegressor(n_estimators=200, random_state=42)

# XGBoost (Hata avcısı - Kaggle şampiyonu)
xgb_model = xgb.XGBRegressor(n_estimators=1000, learning_rate=0.05, n_jobs=-1, random_state=42)

# Linear Regression (Basit ilişkileri garantiye alan sigorta)
lr = LinearRegression()

# 2. SÜPER MODELİ (VOTING) KURUYORUZ
# weights=[1, 2, 1]: XGBoost'un sözü 2 kat değerli olsun dedik.
voting_model = VotingRegressor(
    estimators=[('rf', rf), ('xgb', xgb_model), ('lr', lr)],
    weights=[1, 2, 1] 
)

print("🚀 Süper Model eğitiliyor... (Biraz sürebilir)")
voting_model.fit(X_train, y_train)

# 3. TEST ZAMANI
y_pred_vote = voting_model.predict(X_test)

# Skoru Yazdır
print("-" * 40)
print(f"Süper Model (Voting) R2 Skoru: {r2_score(y_test, y_pred_vote):.4f}")
print("-" * 40)
```

```python
import pandas as pd
import os
import joblib
from sklearn.ensemble import RandomForestRegressor

# 1. ŞAMPİYON MODELİ EĞİT (En iyi ayarlar: Varsayılanlar)
print("🏆 Şampiyon Model (Random Forest) son kez eğitiliyor...")
final_model = RandomForestRegressor(n_estimators=100, random_state=42)
final_model.fit(X_train, y_train)

# 2. GÜVENLİ KAYIT (Klasör yoksa oluşturur)
klasor_yolu = '../models/'  # Proje klasörünün içine kaydeder
os.makedirs(klasor_yolu, exist_ok=True)
dosya_yolu = os.path.join(klasor_yolu, 'house_price_model.pkl')

joblib.dump(final_model, dosya_yolu)
print(f"✅ Model başarıyla kaydedildi: {dosya_yolu}")

# 3. KULLANIM FONKSİYONU (Bunu kopyalayıp saklayabilirsin)
def fiyat_sor(ozellikler):
    # Veriyi hazırla
    df_yeni = pd.DataFrame([ozellikler])
    df_yeni = pd.get_dummies(df_yeni)
    # Eğitimdeki sütunlarla eşitle (Eksikleri 0 yap)
    df_yeni = df_yeni.reindex(columns=X_train.columns, fill_value=0)
    # Tahmin
    tahmin = final_model.predict(df_yeni)
    return tahmin[0]

print("\n🎉 Tebrikler! Yapay Zeka projen tamamlandı.")
print("Artık 'fiyat_sor(benim_evim)' fonksiyonunu kullanarak istediğin evi fiyatlayabilirsin.")
```
